```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Check if the string is equal to its reverse
    if num_str == num_str[::-1]:
        return f"The number {number} is a palindrome."
    else:
        return f"The number {number} is not a palindrome."

# Example usage:
num_to_check = 121
print(is_palindrome(num_to_check))

num_to_check = 123
print(is_palindrome(num_to_check))

num_to_check = 1221
print(is_palindrome(num_to_check))
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"The number {number} is even."
    else:
        return f"The number {number} is odd."

# Example usage:
num_to_check = 10
print(check_odd_even(num_to_check))

num_to_check = 7
print(check_odd_even(num_to_check))
```

# Task
Create an expert system capable of making decisions based on defined rules, exemplified by suggesting activities according to weather conditions. The system should include knowledge representation, an inference engine, and a user interface for interaction.

## Define a Simple Problem Domain and Rules

### Subtask:
Define a simple problem domain for the expert system to make decisions on, such as suggesting an activity based on weather conditions. This will involve outlining the conditions and the corresponding actions/recommendations.


### Problem Domain: Activity Recommendation based on Weather

**Goal:** To recommend a suitable activity based on current weather conditions and temperature.

**1. Input Conditions:**
*   **Weather:**
    *   `'sunny'`
    *   `'rainy'`
    *   `'cloudy'`
    *   `'snowy'`
*   **Temperature:**
    *   `'warm'` (e.g., > 20°C / 68°F)
    *   `'mild'` (e.g., 10-20°C / 50-68°F)
    *   `'cold'` (e.g., < 10°C / 50°F)

**2. Possible Activities/Recommendations:**
*   `'Go for a walk'`
*   `'Read a book'`
*   `'Watch a movie'`
*   `'Build a snowman'`
*   `'Go swimming'`
*   `'Visit a museum'`
*   `'Play board games'`

**3. 'If-Then' Rules:**
1.  **IF** Weather is `'sunny'` **AND** Temperature is `'warm'` **THEN** Suggest `'Go swimming'`
2.  **IF** Weather is `'sunny'` **AND** Temperature is `'mild'` **THEN** Suggest `'Go for a walk'`
3.  **IF** Weather is `'cloudy'` **AND** Temperature is `'mild'` **THEN** Suggest `'Visit a museum'`
4.  **IF** Weather is `'rainy'` **AND** Temperature is `'cold'` **THEN** Suggest `'Watch a movie'`
5.  **IF** Weather is `'rainy'` **AND** Temperature is `'mild'` **THEN** Suggest `'Read a book'`
6.  **IF** Weather is `'snowy'` **AND** Temperature is `'cold'` **THEN** Suggest `'Build a snowman'`
7.  **IF** Weather is `'cloudy'` **AND** Temperature is `'cold'` **THEN** Suggest `'Play board games'`

## Represent Knowledge (Rules)

### Subtask:
Implement the defined rules in a structured way, such as using Python dictionaries or lists of if-else statements, to represent the knowledge base of the expert system.


**Reasoning**:
The subtask requires implementing the defined rules in a structured way using a Python list of dictionaries. I will create a Python list named `rules` and populate it with dictionaries, each representing a rule with 'weather', 'temperature', and 'activity' keys based on common expert system rule structures, then print the list.



```python
rules = [
    {
        "weather": "sunny",
        "temperature": "warm",
        "activity": "Go swimming"
    },
    {
        "weather": "rainy",
        "temperature": "cold",
        "activity": "Read a book"
    },
    {
        "weather": "cloudy",
        "temperature": "mild",
        "activity": "Go for a walk"
    },
    {
        "weather": "snowy",
        "temperature": "cold",
        "activity": "Build a snowman"
    },
    {
        "weather": "sunny",
        "temperature": "cold",
        "activity": "Go ice skating"
    }
]

print("Structured Knowledge Base (Rules):")
for rule in rules:
    print(rule)
```

## Develop an Inference Engine

### Subtask:
Create a function or a set of functions that acts as the inference engine. This engine will take user input (e.g., current weather) and use the knowledge base to deduce and provide a decision or recommendation.


**Reasoning**:
To implement the inference engine, I will define the `infer_activity` function as instructed. This function will iterate through the `rules` list, match user inputs for weather and temperature to the rules, and return a corresponding activity or a default message if no match is found. Finally, I will demonstrate its functionality with example calls.



```python
def infer_activity(current_weather, current_temperature):
    """
    Inference engine to suggest an activity based on weather and temperature.
    """
    for rule in rules:
        if rule["weather"] == current_weather and rule["temperature"] == current_temperature:
            return rule["activity"]
    return "Sorry, I can't suggest an activity for these conditions."

# Example usage to test the inference engine:
print("\nTesting the inference engine:")

# Test cases from the problem domain
activity1 = infer_activity("sunny", "warm")
print(f"Weather: sunny, Temperature: warm -> Activity: {activity1}")

activity2 = infer_activity("rainy", "cold")
print(f"Weather: rainy, Temperature: cold -> Activity: {activity2}")

activity3 = infer_activity("cloudy", "mild")
print(f"Weather: cloudy, Temperature: mild -> Activity: {activity3}")

activity4 = infer_activity("snowy", "cold")
print(f"Weather: snowy, Temperature: cold -> Activity: {activity4}")

# Test case not explicitly in the initial rules but added in previous step
activity5 = infer_activity("sunny", "cold")
print(f"Weather: sunny, Temperature: cold -> Activity: {activity5}")

# Test case for no match
activity_no_match = infer_activity("stormy", "hot")
print(f"Weather: stormy, Temperature: hot -> Activity: {activity_no_match}")

activity_no_match_valid_conditions = infer_activity("cloudy", "warm")
print(f"Weather: cloudy, Temperature: warm -> Activity: {activity_no_match_valid_conditions}")
```

## Create a User Interface for Interaction

### Subtask:
Develop a simple text-based interface that allows the user to input information and receive decisions from the expert system.


**Reasoning**:
I need to implement a user interface that takes input for weather and temperature, calls the inference engine, and displays the recommended activity. This will involve using `input()` for user interaction and `print()` for output.



```python
print("\n--- Expert System for Activity Recommendation ---")

# 1. Prompt the user to enter the current weather
user_weather = input("Enter the current weather (e.g., sunny, rainy, cloudy, snowy): ").lower()

# 2. Prompt the user to enter the current temperature condition
user_temperature = input("Enter the current temperature (e.g., warm, mild, cold): ").lower()

# 3. Call the infer_activity function to get the recommended activity
recommended_activity = infer_activity(user_weather, user_temperature)

# 4. Print the recommended activity to the console
print(f"\nBased on your input:\n  Weather: {user_weather}\n  Temperature: {user_temperature}\n\nRecommendation: {recommended_activity}")
```
